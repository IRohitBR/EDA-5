# INPLACE

- when we do the opertaions the result output is not overwriting.

- means the original dataframe is same as it is.

- in order to do this we need use **inplace**

- by default inplace=False

- make **inplace=True** if you want overwrite the result

```python
import pandas as pd
d={'Names':['Ramesh','Suresh','Satish'],
   'Age':[10,20,30],
    'City':['Hyd','Pune','Blr']}
df=pd.DataFrame(d)
df
```

```python
marks=[20,30,40]
df['marks']=marks
df
```

```python
df=pd.DataFrame()
df.drop('marks',axis=0,inplace=True)
```
